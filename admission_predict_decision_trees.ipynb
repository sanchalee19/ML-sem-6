```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML/ML Dataset/Admission_Predict.csv')
df.head()
```

```python
df = df.drop('Serial No.', axis = True)
```

```python
df.head()
```

```python
std_scaler = StandardScaler()
```

```python
df['GRE Score'] = std_scaler.fit_transform(df['GRE Score'].values.reshape(-1, 1))
df['TOEFL Score'] = std_scaler.fit_transform(df['TOEFL Score'].values.reshape(-1, 1))
```

```python
df.head()
```

```python
df['Chance of Admit '] = (df['Chance of Admit '] > 0.5).astype(int)
```

```python
X = df.drop('Chance of Admit ', axis = True)
y = df['Chance of Admit ']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train
```

```python
y_train
```

## Modelling using GridSearchCV

```python
dtree_clf = DecisionTreeClassifier()
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [6, 8, 10],
    'min_samples_split': [20, 30, 40],
    'random_state': [28, 42]
}
```

```python
grid_search = GridSearchCV(dtree_clf, param_grid=param_grid, scoring='neg_log_loss', cv = 5)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_params
```

```python
best_dtree_clf = DecisionTreeClassifier(**best_params)
```

```python
best_dtree_clf.fit(X_train, y_train)
```

```python
y_pred = best_dtree_clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
