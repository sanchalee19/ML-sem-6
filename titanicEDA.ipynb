## DEALING WITH NULL VALUES

```python
df.isna().sum()
```

### Conclusions, null values are in:
1. age
2. embarked
3. deck

```python
# Replace missing age values with the median age
median_age = df['age'].median()
df['age'].fillna(median_age, inplace=True)
```

```python
# Calculate the mode of the 'Embarked' column
mode_embarked = df['embarked'].mode()[0]

# Replace missing values with the mode
df['embarked'].fillna(mode_embarked, inplace=True)
```

```python
# Calculate the percentage of missing values in the 'Deck' column
missing_percentage = (df['deck'].isnull().sum() / len(df['deck'])) * 100

print(f"Percentage of missing values in 'Deck': {missing_percentage:.2f}%")
```

```python
df.drop(columns = 'deck', inplace = True)
```

```python
df.isna().sum()
```
